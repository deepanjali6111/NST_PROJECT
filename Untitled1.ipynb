```python
def load_image(img_path, max_dim=512):
    """Load and preprocess image"""
    img = tf.io.read_file(img_path)
    img = tf.image.decode_image(img, channels=3)
    img = tf.image.convert_image_dtype(img, tf.float32)

    shape = tf.cast(tf.shape(img)[:-1], tf.float32)
    long_dim = max(shape)
    scale = max_dim / long_dim

    new_shape = tf.cast(shape * scale, tf.int32)
    img = tf.image.resize(img, new_shape)
    return img[tf.newaxis, :]

def deprocess_img(img):
    """Convert tensor to displayable format"""
    img = 255 * img
    return tf.cast(img, tf.uint8)
```

```python
content_layers = ['block5_conv2']
style_layers = [
    'block1_conv1',
    'block2_conv1',
    'block3_conv1',
    'block4_conv1',
    'block5_conv1'
]

def get_vgg_model():
    """Create VGG19 model with access to intermediate layers"""
    vgg = vgg19.VGG19(include_top=False, weights='imagenet')
    vgg.trainable = False
    outputs = [vgg.get_layer(name).output for name in (style_layers + content_layers)]
    return Model(vgg.input, outputs)
```

```python
def content_loss(content, generated):
    """Calculate content loss between features"""
    return tf.reduce_mean(tf.square(content - generated))

def gram_matrix(input_tensor):
    """Calculate Gram matrix for style representation"""
    result = tf.linalg.einsum('bijc,bijd->bcd', input_tensor, input_tensor)
    input_shape = tf.shape(input_tensor)
    num_locations = tf.cast(input_shape[1] * input_shape[2], tf.float32)
    return result / num_locations

def style_loss(style, generated):
    """Calculate style loss using Gram matrices"""
    S = gram_matrix(style)
    G = gram_matrix(generated)
    return tf.reduce_mean(tf.square(S - G))
```

```python
def style_transfer_training(content_image, style_image,
                          epochs=1000,
                          content_weight=5e2,  # Reduced from 1e3
                          style_weight=1e-1,   # Increased from 1e-2
                          tv_weight=1e-4):
    """Main training loop for style transfer with advanced improvements"""
    model = get_vgg_model()

    # Get style and content targets
    content_targets = model(content_image * 255.0)[len(style_layers):]
    style_targets = model(style_image * 255.0)[:len(style_layers)]

    # Initialize with noise-blended content image for better exploration
    noise = tf.random.normal(content_image.shape) * 0.1
    generated_image = tf.Variable(content_image * 0.9 + noise)

    # Learning rate scheduling
    initial_lr = 0.02
    decay_steps = epochs // 3
    lr_schedule = tf.keras.optimizers.schedules.ExponentialDecay(
        initial_learning_rate=initial_lr,
        decay_steps=decay_steps,
        decay_rate=0.5,
        staircase=True
    )

    # Using a more sophisticated optimizer with momentum
    optimizer = tf.optimizers.Adam(learning_rate=lr_schedule, beta_1=0.9, beta_2=0.999)

    # Track best result
    best_loss = float('inf')
    best_image = None

    # Layer weights for style - giving more weight to earlier layers
    style_layer_weights = [1.0, 0.8, 0.6, 0.4, 0.2]

    def total_variation_loss(image):
        x_var = tf.reduce_sum(tf.square(image[:, 1:, :, :] - image[:, :-1, :, :]))
        y_var = tf.reduce_sum(tf.square(image[:, :, 1:, :] - image[:, :, :-1, :]))
        return x_var + y_var

    for epoch in range(epochs):
        current_lr = lr_schedule(epoch)

        with tf.GradientTape() as tape:
            # Get model outputs
            outputs = model(generated_image * 255.0)

            # Split into style and content features
            content_outputs = outputs[len(style_layers):]
            style_outputs = outputs[:len(style_layers)]

            # Calculate content loss
            content_loss_value = content_weight * content_loss(
                content_targets[-1], content_outputs[-1])

            # Calculate weighted style loss
            style_loss_value = 0
            for i, (target, output, weight) in enumerate(zip(style_targets, style_outputs, style_layer_weights)):
                layer_style_loss = style_loss(target, output) * style_weight * weight
                style_loss_value += layer_style_loss

            # Normalize style loss
            style_loss_value /= sum(style_layer_weights)

            # Add total variation loss with weight
            tv_loss = total_variation_loss(generated_image) * tv_weight

            # Total loss
            total_loss = content_loss_value + style_loss_value + tv_loss

        # Apply gradients
        gradients = tape.gradient(total_loss, generated_image)

        # Gradient clipping to prevent extreme updates
        gradients = tf.clip_by_norm(gradients, 10.0)

        optimizer.apply_gradients([(gradients, generated_image)])
        generated_image.assign(tf.clip_by_value(generated_image, 0.0, 1.0))

        # Update display every 100 epochs
        if epoch % 100 == 0:
            clear_output(wait=True)

            # Convert to numpy and handle display properly
            img_array = generated_image.numpy()[0]  # Remove batch dimension
            img_array = np.clip(img_array * 255, 0, 255).astype(np.uint8)

            plt.figure(figsize=(8, 8))
            plt.imshow(img_array)
            plt.title(f'Epoch {epoch}\nLoss: {total_loss:.2f} | LR: {current_lr:.4f}')
            plt.axis('off')
            plt.show()

            # Print detailed loss components
            print(f"Content Loss: {content_loss_value:.2f}")
            print(f"Style Loss: {style_loss_value:.2f}")
            print(f"TV Loss: {tv_loss:.2f}")

        # Track best result
        if total_loss < best_loss:
            best_loss = total_loss
            best_image = generated_image.numpy()

    # Return both best image and final image
    return best_image, generated_image.numpy(), best_loss

# Update the deprocess_img function globally as well
def deprocess_img(img):
    """Convert tensor to displayable format"""
    # Convert to numpy if tensor
    if isinstance(img, tf.Tensor):
        img = img.numpy()

    # Scale to 0-255
    img = np.clip(img * 255, 0, 255).astype(np.uint8)

    # Remove batch dimension if it exists
    if len(img.shape) == 4:
        img = img[0]

    return img
```

```python
# File upload widgets
content_upload = FileUpload(description="Content Image", accept='.jpg,.png')
style_upload = FileUpload(description="Style Image", accept='.jpg,.png')

# Parameter sliders
style_weight_slider = FloatSlider(
    value=1e-2, min=1e-4, max=1e-1, step=1e-4,
    description='Style Weight', readout_format='.4f'
)
content_weight_slider = FloatSlider(
    value=1e3, min=1e2, max=1e4, step=100,
    description='Content Weight', readout_format='.0f'
)
epochs_slider = FloatSlider(
    value=1000, min=500, max=3000, step=100,
    description='Epochs', readout_format='.0f'
)

# Process button
process_btn = Button(description="Generate Style Transfer", button_style='success')

def on_process_click(b):
    """Handle generate button click"""
    try:
        # Validate uploads
        if not content_upload.data or not style_upload.data:
            print("â— Please upload both images first!")
            return

        # Save uploaded files
        with open('content.jpg', 'wb') as f:
            f.write(content_upload.data[0])
        with open('style.jpg', 'wb') as f:
            f.write(style_upload.data[0])

        # Load and display images
        content_image = load_image('content.jpg')
        style_image = load_image('style.jpg')

        # Show upload preview
        plt.figure(figsize=(10, 5))
        plt.subplot(1, 2, 1)
        plt.imshow(deprocess_img(content_image[0]))
        plt.title('Content Image')
        plt.axis('off')

        plt.subplot(1, 2, 2)
        plt.imshow(deprocess_img(style_image[0]))
        plt.title('Style Image')
        plt.axis('off')
        plt.show()

        # Start style transfer
        print("ðŸš€ Starting style transfer...")
        result = style_transfer_training(
            content_image,
            style_image,
            epochs=int(epochs_slider.value),
            content_weight=content_weight_slider.value,
            style_weight=style_weight_slider.value
        )

        # Show final result
        plt.figure(figsize=(10, 5))
        plt.subplot(1, 2, 1)
        plt.imshow(deprocess_img(content_image[0]))
        plt.title('Original Content')
        plt.axis('off')

        plt.subplot(1, 2, 2)
        plt.imshow(deprocess_img(result[0]))
        plt.title('Styled Result')
        plt.axis('off')
        plt.show()

    except Exception as e:
        print(f"âŒ Error: {str(e)}")

# Connect button to handler
process_btn.on_click(on_process_click)

# Arrange UI components
controls = VBox([
    HBox([content_upload, style_upload]),
    style_weight_slider,
    content_weight_slider,
    epochs_slider,
    process_btn
])

# Display UI
display(controls)



```
